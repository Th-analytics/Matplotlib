```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from scipy import optimize
%matplotlib inline
```

## We have the min and max temperatures in a city In India for each months of the year.We would like to find a function to describe this and show it graphically, the datasetgiven below.

### Task:
1.fitting it to the periodic function

2.plot the fit
#### Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
#Function to get the points of the curve
def calculate(mon,a,b):
    return (b+a*np.cos(mon*2*np.pi/mon.max()))
#Initialization
months = np.arange(12)
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
#Calculation
avg_max = sum(Max)/12
avg_min = sum(Min)/12
amplitude_max = (max(Max)-min(Max))/2
amplitude_min = (max(Min)-min(Min))/2
#Calling The scipy optimize.curve funtion
max_params, params_covariance = optimize.curve_fit(calculate, months,Max, [avg_max,amplitude_max])
min_params, params_covariance = optimize.curve_fit(calculate, months, Min,[avg_min,amplitude_min])
#ploting
plt.scatter(months,Max,color = 'r')
plt.plot(months,calculate(months,*max_params),color = 'r')
plt.scatter(months,Min,color='b')
plt.plot(months,calculate(months,*min_params),color= 'b')
#Labeling
plt.xlabel('Months')
plt.ylabel('Temperature Â°C')
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
data = pd.read_csv(url)
```

#### 1. Create a pie chart presenting the male/female proportion

```python
def genderC(g_data):
    male= 0
    female = 0
    others = 0 
    for i in g_data:
        if i == 'female':
            female+=1
        elif i == 'male':
            male+=1
        else:
            others+=1
    return male,female,others
gender = genderC(data['sex'])
print(gender)
fig, ax = plt.subplots()
a = ax.pie(x=gender,labels = ['Male','Female','Not Mentioned'],colors=['#ff0000','#ffbf00','#ffc9bf'],autopct = '%.2f')
```

#### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender 

```python
groups = data.groupby('sex')
for name, group in groups:
    plt.plot(group["age"], group["fare"], marker="^", linestyle="", label=name)
plt.legend()
```
